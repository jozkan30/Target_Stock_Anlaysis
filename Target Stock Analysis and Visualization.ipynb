```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt


tgt = yf.Ticker('TGT')
tgt_history = tgt.history(period='max')
print(tgt_history)
tgt_history.dropna(inplace=True)
tgt_history.reset_index(inplace=True)
tgt_history["Date"] = pd.to_datetime(tgt_history["Date"]).dt.date

tgt_history["Percent Change"] = tgt_history["Close"].pct_change()
mean_price = tgt_history["Close"].mean()
std_price = tgt_history["Close"].std()

print(round(mean_price, 2))
print(round(std_price, 2))
```

```python
plt.plot(tgt_history["Date"], tgt_history["Close"])
plt.title("TGT Stock Price Over Time")
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()

plt.hist(tgt_history["Percent Change"], bins=50)
plt.title("Distribution of Daily Percent Changes")
plt.xlabel("Percent Change")
plt.ylabel("Frequency")
plt.show()
```
